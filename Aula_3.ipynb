```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
```

```python
plt.plot(data.sepallength, data.sepalwidth, ls='', marker='o', label='sepal')
plt.plot(data.petallength, data.petalwidth, ls='', marker='o', label='petal')
```

```python
data.columns
```

```python
plt.hist(data.sepallength, bins=25)
```

```python
(data.groupby('class').mean().plot(color=['red','blue','black','green'], fontsize=10.0, figsize=(4,4)))
```

```python
import seaborn as sns
```

```python
sns.jointplot(x='sepallength',y='sepalwidth', data=data, height=4)
```
